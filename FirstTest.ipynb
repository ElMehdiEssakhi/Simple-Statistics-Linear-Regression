# Linear Regression - Quick Summary
This notebook performs a simple linear regression on the dataset in `data.csv` (100 rows, columns `x` and `y`) using the functions from `relations.py`. It computes basic statistics and plots a scatter with the fitted regression line.


```python
import pandas as pd
import relations as rl
import matplotlib.pyplot as plt
```

1. Load the data using pandas

```python
data=pd.read_csv("data.csv")
```

2. Assign `x` and `y` (cell that sets `x = data["x"]`, `y = data["y"]`)  

```python
x=data["x"]
y=data["y"]
```

3. Visualize scatter (`plt.scatter(x, y)`)  

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dataset')
```

4. Compute statistics and regression (`rl.moyenne`, `rl.variance`, `rl.covariance`, `rl.a_b`)  

```python
print("moyenne(x):", rl.moyenne(x))
print("moyenne(y):", rl.moyenne(y))
print("variance(x):", rl.variance(x))
print("variance(y):", rl.variance(y))
print("covariance(x, y):", rl.covariance(x, y))
print("correlation(x, y):", rl.correlation(x, y))
a=rl.a_b(x, y)[0]
b=rl.a_b(x, y)[1]
print("Droite de régression: y =", a, "* x +", b)

```

5. Plot fitted line (`plt.plot(x, a*x + b, color='red')`)

```python
plt.Figure()
plt.scatter(x,y)
plt.plot(x, a*x + b, color='red')
```
